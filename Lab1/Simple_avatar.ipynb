```python
import sys
import numpy as np
from matplotlib import pyplot as plt
```

```python
def hex2rgb(value):
    value = value.lstrip('#')
    lv = len(value)
    return tuple(int(value[i:i + lv // 3], 16) for i in range(0, lv, lv // 3))


def rgb2hex(rgb):
    return '#%02x%02x%02x' % rgb
```

```python
cmaps = [
    "#000000",
    '#FF0000',
    "#00FF00",
    "#0CAFEB",
    "#F5D718",
    "#EB63E2",
    "#A1FF4A",
    "#EEEEEE"
]
```

```python
im0=[]
for i in cmaps:
  im0.append(hex2rgb(i))
im0=np.repeat(np.expand_dims(im0, axis=0), repeats=8, axis=0)  
plt.imshow(im0)
```

```python
pvals=[5,5,10,20,5,5,5,5]
pvals=np.asarray(pvals)/float(sum(pvals))
n=len(cmaps)

colors=np.empty([8, 8])
for j in range(8):
    
    colors[j]=np.random.multinomial(n, pvals, size=None)
print(colors)
```

```python
pal=[]
for i in colors:
 ind=0
 for j in i:
   for t in range(int(j)):
     pal.append(hex2rgb(cmaps[int(ind)]))
   ind+=1
im=np.asarray(pal).reshape(8,8,3)
print(pal)
```
